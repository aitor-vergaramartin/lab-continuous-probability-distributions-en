## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
uniform.rvs(2, 8, size=10)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_uniform(bottom, ceiling, count):
    return uniform.rvs(
        loc=bottom,
        scale=ceiling - bottom,
        size=count
    )
```

```python
# your code here
set_1 = generate_uniform(10, 15, 100)
set_2 = generate_uniform(10, 60, 1000)
```

```python
# your code here
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(set_1, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Set 1")

plt.subplot(1, 2, 2)
plt.hist(set_2, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Set 2")

plt.tight_layout()
plt.show()
```

```python
#With shared y axis
fig, axes = plt.subplots(1, 2, figsize=(10, 4),sharey=True)

axes[0].hist(set_1, bins=10)
axes[0].set_title("Set 1")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Count")

axes[1].hist(set_2, bins=10)
axes[1].set_title("Set 2")
axes[1].set_xlabel("Value")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below
# Both distributions are uniform, but Set 2 appears smoother and more evenly distributed due to its larger sample size. 
# Set 1 shows more variability because it contains fewer observations. 
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def generate_expo(size, mean=10):
    values = np.random.exponential(scale=mean, size=size)
    return values.tolist()
```

```python
set_5 = generate_expo(10,10)
set_6 = generate_expo(100,10)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(set_5, bins=100)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Set 5")

plt.subplot(1, 2, 2)
plt.hist(set_6, bins=100)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Set 6")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
#Set 5 (the one with less samples) doesn't have enough data to show the exponential distribution, while set 6 has more low values to compensate the high extreme values, and keep the same mean
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
from scipy.stats import expon

mean = 10
t = 15

probability = expon.cdf(t, scale=mean)
print(f"Probability the customer spends less than 15 minutes: {probability:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
probability2 = 1-probability
print(f"Probability the customer spends more than 15 minutes: {probability2:.4f}")
```
